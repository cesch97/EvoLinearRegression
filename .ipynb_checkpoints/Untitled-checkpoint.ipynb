```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df = pd.read_csv('data/wine-quality/winequality-white.csv', delimiter=';', header=0)
df
```

```python
hist = df.hist()
```

```python
X_df = df.loc[:, df.columns != 'quality']
X_df = sm.add_constant(X_df)
y = df['quality']
```

```python
ols = sm.OLS(y, X_df)
model = ols.fit()
model.summary()
```
